```python
import pandas as pd

df = pd.read_csv('./final.csv', index_col = 0)
df.head(20)
```

```python
import urllib.parse

base_url = 'https://www.youtube.com/results?search_query='
link_list = []

for i in range(len(df)):
    title = df['title'].iloc[i]
    artist = df['artist'].iloc[i]

    # URL-encoded string으로 변환
    encoded_title = urllib.parse.quote_plus(title)  # ' ' -> '+' / '#' -> '%23' / etc.
    encoded_artist = urllib.parse.quote_plus(artist)

    # 하나의 링크로 만들기
    link = base_url + encoded_title + "+" + encoded_artist

    link_list.append(link)

df['youtube search link'] = link_list
```

# Bring first video's link

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import time

# Set up the WebDriver
from selenium.common.exceptions import NoSuchElementException, WebDriverException

def get_first_video_link(search_url):
    options = Options()
    # options.add_argument('headless')
    options.add_argument('--start-maximized')
    options.add_experimental_option('detach', True)
    
    driver = webdriver.Chrome(options=options)

    try:
        # Navigate to the provided URL
        driver.get(search_url)
        time.sleep(1)

        elems = driver.find_elements(By.CSS_SELECTOR, '#thumbnail')
        if elems and len(elems) > 1:
            video_link = elems[1].get_attribute('href')
        else:
            video_link = search_url  # Default to the search URL if the video link is not found
    except (NoSuchElementException, WebDriverException) as e:
        print(f"Error encountered for URL {search_url}: {str(e)}")
        video_link = 'None'  # Default to the search URL in case of any error
    finally:
        # Close the browser window
        driver.quit()

    return video_link
```

```python
get_first_video_link('https://www.youtube.com/results?search_query=shut+down')
```

```python
link_list = []

for i in range(len(df)):
    link = get_first_video_link(df['youtube search link'].iloc[i])
    link_list.append(link)
    
df['first video link'] = link_list
```
