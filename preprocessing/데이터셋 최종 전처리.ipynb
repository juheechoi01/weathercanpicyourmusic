```python
import pandas as pd
import numpy as np
```

```python
melon = pd.read_csv('playlist.csv')
genie = pd.read_csv('playlist_중복처리.csv')
genie.drop('Unnamed: 0', axis=1, inplace=True)

#문자열에서 리스트로 변환
for i in range(len(melon)):
    melon['specific_tag'][i] = eval(melon['specific_tag'][i])

# Number가 1 그리고 그 외로 분리
melon_1 = melon[melon['number']==1]
melon_1.drop('tag', axis=1, inplace=True)
melon_2 = melon[melon['number']!=1]
melon_2.drop('tag', axis=1, inplace=True)
```

```python
ex_2 = melon_2.head(10)

total = []
small = []
for i in range(len(ex_2)-1):
    num1 = ex_2['number'][i]
    num2 = ex_2['number'][i+1]
    if num1 == num2:
        for j in ex_2['specific_tag'][i]:
            if j in small:
                continue
            else:
                small.append(j)
    else:
        for j in ex_2['specific_tag'][i]:
            if j in small:
                continue
            else:
                small.append(j)
        total.append(small)
        small = []

for k in ex_2['specific_tag'][-1]:
    if k in small:
        continue
    else:
        small.append(k)
total.append(small)

total
```
